```python
from simlib.VoteSim import VotingSimulation
from simlib.VoteSim import random_walk, move_towards_winner, cluster_attraction

import numpy as np
```

```python
sim = VotingSimulation(num_voters=200,
                       num_candidates=4,
                       center=np.array([0, 0]),
                       spread_radius=15)
```

```python
sim.vote()
sim.get_result()
sim.plot() 
```

```python
sim.execute(random_walk, step_size=2.0)
```

```python
sim.vote()
sim.get_result()
sim.plot()
```

```python
sim.execute(move_towards_winner, fraction=0.2)
```

```python
sim.vote()
sim.get_result()
sim.plot()
```
