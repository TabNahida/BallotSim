```python
from simlib.VoteSim import VotingSimulation
from simlib.VoteSim import evolve_add_noise, evolve_towards_voter_center, evolve_away_from_voter_center, candidate_random_walk, winner_move_away_from_voter_center, winner_random_walk
from simlib.VoteSim import random_walk, move_towards_winner, cluster_attraction, probabilistic_move_towards_winner

import numpy as np
```

```python
sim = VotingSimulation(num_voters=200,
                       num_candidates=3,
                       center=np.array([0, 0]),
                       spread_radius=15)
```

```python
sim.vote()
sim.get_result()
sim.plot()
```

```python
for i in range(5):
    for j in range(5):
        sim.execute(probabilistic_move_towards_winner, fraction=0.2)
        sim.propose(winner_random_walk, step_size=2.0)
    sim.propose(winner_move_away_from_voter_center, fraction=0.2)
    for j in range(5):
        sim.execute(probabilistic_move_towards_winner, fraction=0.2)
        sim.propose(winner_random_walk, step_size=2.0)
    sim.vote()
    sim.get_result()
    sim.plot()
```
