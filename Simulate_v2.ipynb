```python
from simlib.VoteSim import VotingSimulation
from simlib.VoteSim import evolve_add_noise, evolve_towards_voter_center, candidate_random_walk
from simlib.VoteSim import random_walk, move_towards_winner, cluster_attraction

import numpy as np
```

```python
sim = VotingSimulation(num_voters=200,
                       num_candidates=3,
                       center=np.array([0, 0]),
                       spread_radius=15)
```

```python
sim.vote()
sim.get_result()
sim.plot() 
```

```python
sim.execute(random_walk, step_size=5.0)
sim.vote()
sim.get_result()
sim.plot()
```

```python
sim.execute(move_towards_winner, fraction=0.5)
sim.vote()
sim.get_result()
sim.plot()
```

```python
sim.execute(cluster_attraction, cluster_pct=0.5)
sim.vote()
sim.get_result()
sim.plot()
```

```python
sim.propose(candidate_random_walk, step_size=2.0)
sim.vote()
sim.get_result()
sim.plot()
```

```python
sim.propose(evolve_towards_voter_center, fraction=0.5)
sim.vote()
sim.get_result()
sim.plot()
```

```python
sim.propose(evolve_add_noise, noise_std=0.5)
sim.vote()
sim.get_result()
sim.plot()
```
