```python
from simlib.VoteSim import VotingSimulation
from simlib.VoteSim import random_walk, move_towards_winner, cluster_attraction
from simlib.VoteSim import evolve_add_noise, evolve_towards_voter_center
```

```python
sim = VotingSimulation(num_voters=200, num_candidates=3, spread_radius=5)
```

```python
sim.vote()
sim.get_result()
sim.plot()
```

```python
sim.propose(new_num_candidates=4, mode='random')
sim.vote()
sim.get_result()
sim.plot()
```

```python
sim.propose(mode='evolve', evolve_func=evolve_towards_voter_center, fraction=0.3)
sim.vote()
sim.get_result()
sim.plot()
```

```python
sim.propose(mode='evolve', evolve_func=evolve_add_noise, noise_std=0.5)
sim.vote()
sim.get_result()
sim.plot()
```
